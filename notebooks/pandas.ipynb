```python
import pandas as pd
import numpy as np
# Create sample data
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [25, 30, 35, 40, 25],
    'salary': [50000, 60000, 75000, 90000, 55000],
    'department': ['HR', 'IT', 'IT', 'Finance', 'HR']
})

# DataFrame: Filter people older than 30 with salary over 70000
result_df = df[(df['age'] > 30) & (df['salary'] > 70000)]
print(result_df)

avg_by_dept_df = df.groupby('department')['salary'].mean()
print(avg_by_dept_df)

# Create a second DataFrame
df2 = pd.DataFrame({
    'department': ['HR', 'IT', 'Finance', 'Marketing'],
    'location': ['New York', 'San Francisco', 'Chicago', 'Austin']
})

# DataFrame: Join employee data with department locations
merged_df = df.merge(df2, on='department')
print(merged_df)

df_missing = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [9, 10, 11, 12]
})

# DataFrame: Fill missing values with column means
filled_df = df_missing.fillna(df_missing.mean())
print(filled_df)

dates = pd.date_range('2023-01-01', periods=5)
ts_df = pd.DataFrame({
    'date': dates,
    'values': [10, 12, 15, 14, 18]
}).set_index('date')

# DataFrame: Resample to monthly average
monthly_df = ts_df.resample('ME').mean()
print(monthly_df)
```
