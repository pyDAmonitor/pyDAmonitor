```python
import numpy as np
import xarray as xr
import holoviews as hv
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm
import hvplot.xarray  # Enables xarray to use hvplot directly

hv.extension('bokeh')

# Example data
data = xr.DataArray(np.random.uniform(-16, 48.4, size=(100, 100)), dims=('lat', 'lon'))

# Set colormap limits
vmin = float(data.min())
vmax = float(data.max())

# Create a TwoSlopeNorm so that 0 is the midpoint
norm = TwoSlopeNorm(vmin=vmin, vcenter=0, vmax=vmax)

# Plot using hvplot with custom normalization and colormap
plot = data.hvplot.image(
    cmap='coolwarm', 
    clim=(vmin, vmax), 
    norm=norm,
    colorbar=True,
    title='Cool below zero, warm above'
)

plot
```

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm
%matplotlib inline


# Example data
lon = np.linspace(0, 10, 100)
lat = np.linspace(0, 10, 100)
X, Y = np.meshgrid(lon, lat)
Z = 10 * np.sin(X) * np.cos(Y)

# Create DataArray
data = xr.DataArray(Z, coords=[('lat', lat), ('lon', lon)])

# Set up norm centered at 0
vmin = float(data.min())
vmax = float(data.max())
norm = TwoSlopeNorm(vmin=-5, vcenter=0, vmax=10)

# Plot using xarray (matplotlib under the hood)
plt.figure(figsize=(8, 6))
contour = data.plot.contourf(
    levels=20,
    cmap='coolwarm',
    norm=norm,
    add_colorbar=True
)

plt.title("Contourf with coolwarm centered at 0")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
import numpy as np
import xarray as xr
import holoviews as hv
from holoviews.operation import contours
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

hv.extension('bokeh')  # You can also use 'matplotlib', but Bokeh works fine here

# Create some sample data that crosses zero
lon = np.linspace(0, 10, 100)
lat = np.linspace(0, 10, 100)
X, Y = np.meshgrid(lon, lat)
Z = 10 * np.sin(X) * np.cos(Y)

# Wrap in xarray DataArray
data = xr.DataArray(Z, coords=[('lat', lat), ('lon', lon)])

# === Step 1: Define symmetric contour levels ===
vabs = max(abs(float(data.min())), abs(float(data.max())))
levels = np.linspace(-5, 10, 20)  # 20 filled bands

# === Step 2: Create a custom ListedColormap ===
base_cmap = plt.get_cmap('coolwarm')
custom_colors = base_cmap(np.linspace(0, 1, len(levels)-1))  # len(levels)-1 segments
custom_cmap = ListedColormap(custom_colors)

# === Step 3: Apply holoviews.operation.contours ===
filled_contours = contours(
    data.plot(),  # convert to Image first
    levels=levels,
    filled=True
)

# === Step 4: Apply styling ===
contour_plot = filled_contours.opts(
    cmap=custom_cmap,
    colorbar=True,
    line_color=None,
    width=600,
    height=500,
    clim=(-vabs, vabs),
    cnorm='linear',  # must be 'linear' since Holoviews doesn’t support TwoSlopeNorm
    tools=['hover'],
    title='Contours with custom coolwarm (0-centered)'
)

contour_plot
```

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm

# Sample data
lon = np.linspace(0, 10, 100)
lat = np.linspace(0, 10, 100)
X, Y = np.meshgrid(lon, lat)
Z = 10 * np.sin(X) * np.cos(Y)
data = xr.DataArray(Z, coords=[('lat', lat), ('lon', lon)])

# Set desired levels
cmin = float(data.min())
cmax = float(data.max())

# Round to nearest nice symmetric range if you want
vabs = max(abs(cmin), abs(cmax))
levels = np.linspace(-5, 10, 16)  # Ensure 0 is in the middle

# Create a colormap with 20 intervals (len(levels)-1)
base_cmap = plt.get_cmap('coolwarm')
custom_cmap = ListedColormap(base_cmap(np.linspace(0.25, 1, len(levels) - 1)))

# Create a norm so each level maps to a color bin
norm = BoundaryNorm(boundaries=levels, ncolors=len(custom_cmap.colors))

# Plot
plt.figure(figsize=(8, 6))
contour = data.plot.contourf(
    levels=levels,
    cmap=custom_cmap,
    norm=norm,
    add_colorbar=True
)

plt.title('Custom coolwarm with 0 at center')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
