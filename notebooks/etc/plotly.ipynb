```python
%%time
import plotly.graph_objects as go
import numpy as np

fig = go.Figure()

for i in range(60):
    x = np.linspace(0, 10, 100)
    y = np.sin(x + i)
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name=f'Plot {i+1}'))

fig.update_layout(title="Interactive Plotly Graphs", showlegend=True)
fig.show()
```
